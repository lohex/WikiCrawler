```python
from WikiCrawler import KnowledgeNet

BiologyKnowledgbase = KnowledgeNet()
BiologyKnowledgbase.setSkipRule([
    ".* about .*",".*logists.*",".*chemists.*",".*People with.*",".*Deaths from.*",".*Living people.*",".*Books with.*"
    ".*stub.*",".*journals.*",".*Society.*","Template:.*","Books about.*","missing person cases",
    ".*journals?",".*Society",".*by country",".*organizations.*","\d{4} births","\d{4} deaths"])

categories = """Biochemistry
Immunology
Biomedicine
Chemical biology
Subfields_of_evolutionary_biology
Branches_of_genetics
Chronobiology
Computational biology
Developmental biology
Evolutionary biology
Genetics
Human biology
Branches of immunology
Microbiology
Molecular biology
Structural biology
Systems biology
Eukaryote biology
Chromosomes (human)
Human genes
Human haplotypes
Prokaryotes
Cell biology
Viruses""".split('\n')

excluded_categories = [
    'Category:Evolutionary psychology',
    'Category:Assistance',
    'Category:Sir William Dunn Professors of Biochemistry',
    'Category:Chemical weapons',
    'Category:Paleontology by year'
    ]

for this_category in categories:
    BiologyKnowledgbase.startScan(start_at=f'Category:{this_category}',depth=5,skip=excluded_categories)
    BiologyKnowledgbase.save('BiologyKnowledgbase.pkl',overwrite=True)
    
BiologyKnowledgbase.indexInfo() 
```

```python
def try_again(fkt):
    def wrapper(*args,**kwargs):
        n_repeats = kwargs['n_repeats'] if 'n_repeats' in kwargs.keys() else 3
        del kwargs['n_repeats']
        success = False
        for k in range(n_repeats):
            try:
                result = fkt(*args,**kwargs)
                success = True
                break
            except:
                print(f'attempt {k+1} failed')
            
        if not success:
            result = -1

        return result
    return wrapper



global counter
counter = 0
@try_again
def this_func(what):
    global counter
    if counter == 3:
        print(f'HERE YOU GO: {what}')
        return 1
    else:
        print(f'{counter}: Not yet ;)')
        counter += 1
        raise Exception("Not at the first trial!")
```

```python
categorial_labels = {}
for article in BiologyKnowledgbase.pages:
    categories = BiologyKnowledgbase.retrieveCategories(article)
    for this_category in categories:
        this_category = this_category.replace('_',' ').replace('Category:','')
        if not this_category in categorial_labels.keys():
            categorial_labels[this_category] = []
        if not article in categorial_labels[this_category]:
            categorial_labels[this_category].append(article)
```

```python
threshold = 100
use_categories = {cat:[len(cnt),3000-len(cnt)] for cat,cnt in sorted(categorial_labels.items(),key=lambda x:-len(x[1])) if len(cnt) >= threshold}
print(len(use_categories))
for cat,counts in use_categories.items():
    print(cat,counts)
```

```python
import re
import os

search_for = ".*ists.*"
hits = {}

blacklist = ['Articles covered','Articles tagged','All articles','All Wikipedia articles','Articles with','.*Wikidata','Articles lacking','Articles needing','Articles containing','Articles requiring','Wikipedia articles in need of']

label_list = {}
for file in os.scandir('BiologyKnowledgbase'):
    with open(file.path,'r') as fp:
        lines = fp.readlines()
    
    labels = set(use_categories).intersection(lines[0].split(', '))
    if len(labels) > 0:
        groups = lines[1].strip().split(', ')
        groups = [g for g in groups if not any([re.match(f"{rule}.*",g) for rule in blacklist])]
    
        for grp in groups:
            if not grp in label_list.keys():
                label_list[grp] = []
            label_list[grp].append(file.name)

        if any([g for g in groups if re.match(search_for,g)]):
            hits[file.name] = groups

        print(file.name,len(labels),groups) 
        #for ln in lines[2:]:
        #    print('\t',ln)
```
